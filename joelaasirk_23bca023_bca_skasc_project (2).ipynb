```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd #Import pandas to use DataFrames

# Load the DataFrame (if not already loaded)
matches_data = pd.read_csv('international_matches.csv')  # Replace with your file path

# Define the column for the ECDF plot
column_name = 'home_team_fifa_rank'  # Replace with the column you want to plot

# Drop missing values
data = matches_data[column_name].dropna()

# Calculate ECDF
x = np.sort(data)
y = np.arange(1, len(x) + 1) / len(x)

# Create the ECDF plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='.', linestyle='none', color='blue', alpha=0.7)

# Customize the plot
plt.title(f'ECDF of {column_name}', fontsize=14)
plt.xlabel(column_name, fontsize=12)
plt.ylabel('ECDF', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('international_matches.csv') #Changed file path to the current working directory

# Select numerical columns for correlation analysis
numerical_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_columns.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Water Intake Tracker
# Simulate water intake data
water_intake = pd.DataFrame({
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'liters': [2.0, 2.5, 3.0, 2.8, 2.2, 3.1, 2.7]
})

# --- Improvements ---

# 1. Highlight target intake (e.g., 2.5 liters)
target_intake = 2.5

# 2. Make it more visually appealing
colors = ['skyblue' if liters >= target_intake else 'lightcoral' for liters in water_intake['liters']]

# --- Create the plot ---

plt.figure(figsize=(10, 6))
plt.bar(water_intake['day'], water_intake['liters'], color=colors, alpha=0.7, edgecolor='black')
plt.title('Weekly Water Intake')
plt.xlabel('Day')
plt.ylabel('Liters of Water')

# Add target intake line
plt.axhline(y=target_intake, color='gray', linestyle='--', label=f'Target: {target_intake} liters')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()  # Show legend for target line
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Water Intake Tracker
water_intake = pd.DataFrame({
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'liters': [2.0, 2.5, 3.0, 2.8, 2.2, 3.1, 2.7]
})

# --- Improvements ---

# 1. Add target intake line
target_intake = 2.5

# 2. Customize appearance
boxplot_kwargs = {
    'color': 'skyblue',
    'width': 0.4,  # Adjust box width
    'fliersize': 5,  # Adjust outlier marker size
    'notch': True  # Add a notch to the box
}

# --- Create the plot ---

plt.figure(figsize=(8, 6))
sns.boxplot(y=water_intake['liters'], **boxplot_kwargs)  # Apply custom settings
plt.title('Boxplot of Weekly Water Intake Distribution')  # Improved title
plt.ylabel('Liters of Water')

# Add target intake line
plt.axhline(y=target_intake, color='gray', linestyle='--', label=f'Target: {target_intake} liters')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()  # Show legend for target line
plt.show()
```
